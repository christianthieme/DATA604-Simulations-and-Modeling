```python
#This is the code for x amount of chickens for one day

day_count = []
egg = [0,1]

chance_for_egg = np.random.uniform(low = 235, high = 275, size = 6)

print("Samples from uniform distribution: {}\n".format(chance_for_egg))

for i in chance_for_egg:
    probabilities = [1-(i/365),(i/365)]
    outcome = np.random.choice(egg, size = 1, p= probabilities)
    #print(round(i)/365)
    day_count.append(outcome)
    print("Outcome is {}".format(outcome))
print("Total count is {}".format(sum(day_count)))
```

```python
total_egg_count = []

for day in range(31):
    day_count = []
    egg = [0,1]

    chance_for_egg = np.random.uniform(low = 235, high = 275, size = 25)
    
    for i in chance_for_egg:
        probabilities = [1-(i/365),(i/365)]
        outcome = np.random.choice(egg, size = 1, p= probabilities)
        day_count.append(outcome[0])
        print("Day {} outcome is {}".format(day + 1, outcome[0]))
    total_egg_count.append(sum(day_count))
    
for egg in range(len(total_egg_count)): 
    print("\nDay {} egg count:{}".format(egg+1, total_egg_count[egg]))

print("\nTotal egg count: {}".format(sum(total_egg_count)))
```
